```python
import pandas as pd
```

```python
data = pd.read_csv('NYC_Property_Data.csv')
```

```python
data.isnull().sum()
```

```python
data.dropna(how = 'any', inplace = True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
features = data.drop(['Unnamed: 0', 'Unnamed: 0.1','COMMERCIAL UNITS', 'ADDRESS', 'SALE DATE', 'SALE PRICE'], axis = 1)
features.head()
```

```python
target = data['SALE PRICE']
target.head()
```

```python
lowprices= target.query('(SALE PRICE > 0) & (SALE PRICE < 20.0)')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
features['NEIGHBORHOOD',] = le.fit_transform(features['NEIGHBORHOOD']) # converts all the categorical data to numeric and reflects the changes in the list
```

```python
features['BUILDING CLASS CATEGORY'] = le.fit_transform(features['BUILDING CLASS CATEGORY']) # converts all the categorical data to numeric and reflects the changes in the list
```

```python
features['TAX CLASS AT PRESENT'] = le.fit_transform(features['TAX CLASS AT PRESENT'])
features['BUILDING CLASS AT PRESENT'] = le.fit_transform(features['BUILDING CLASS AT PRESENT'])
features['BUILDING CLASS AT TIME OF SALE'] = le.fit_transform(features['BUILDING CLASS AT TIME OF SALE'])
```

```python
features
```

```python
from sklearn.model_selection import train_test_split #to split test and train
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(features_train, target_train)
```

```python
target_pred = model.predict(features_test)
```

```python
pd.DataFrame({'Actual': target_test, 'Predicted': target_pred, 'Error': target_test-target_pred})
```
